```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd

from matplotlib.offsetbox import (TextArea, DrawingArea, OffsetImage,
                                  AnnotationBbox)
from matplotlib.cbook import get_sample_data
import matplotlib.image as img
```

```python


fig, ax = plt.subplots()

# Define a 1st position to annotate (display it with a marker)
xy = (0.5, 0.8)
ax.plot(xy[0], xy[1], ".r")

arr_img= img.imread("home.png")

imagebox = OffsetImage(arr_img, zoom=1)
imagebox.image.axes = ax

ab = AnnotationBbox(imagebox, xy,
                    xybox=(0.,  -30.),
                    xycoords='data',
                    boxcoords="offset points",
                    pad=0,
                    frameon = False,
                    arrowprops=dict(
                        arrowstyle="->",
                        connectionstyle="angle,angleA=10,angleB=90,rad=15")
                    )
ax.add_artist(ab)
plt.show()
```

# Dessinner un nombre de petits dessins proportionnel aux nombres
Au passage on ajoute une légende.

```python
import matplotlib.pyplot as plt
import pandas as pd

fig = plt.figure(figsize=(10, 6), facecolor='white')
ax = fig.add_subplot(1, 1, 1)
# Les deux lignes ci dessus permettent plus d'options que le simple
# fig, ax = plt.subplots()

work_df = sub_epi_df
plt.bar(work_df.event, work_df.deaths)

arr_img= img.imread("home.png")


units_per_icon = 5000000
taille = 1

# imagebox.image.axes = ax
for x,y in zip(work_df.event, work_df.deaths):
    print("x, y", x, y )
    nb_of_icons = int(round(y / units_per_icon))
    print("nb_incones", nb_of_icons)
    for i in range(nb_of_icons):

        imagebox = OffsetImage(arr_img, zoom=0.7* math.sqrt(taille))
        ab = AnnotationBbox(imagebox, (x, int(y/ nb_of_icons) * i + (y/nb_of_icons/2)),
                            xybox=(0,  0),
                            xycoords='data',  # on s'exprime en coordonnées de data
                            boxcoords="offset points",  # le décalage est en points
                            frameon=False,  # on élimine le cadre autour du dessin.
                            pad=1,
                        # arrowprops=dict(
                        #     arrowstyle="->",
                        #     connectionstyle="angle,angleA=0,angleB=90,rad=3")
                            )
        ax.add_artist(ab)
plt.legend(["1 M Morts"], frameon = False)
plt.show()
```

La légende serait plus belle avec un dessin ! J'ai trouvé [sur cette page](https://stackoverflow.com/questions/26029592/insert-image-in-matplotlib-legend) une solution. J'ai juste à adapter.

```python
import os

from matplotlib.transforms import TransformedBbox
from matplotlib.image import BboxImage, Bbox
from matplotlib.legend_handler import HandlerBase
# from matplotlib._png import read_png

class ImageHandler(HandlerBase):
    def create_artists(self, legend, orig_handle,
                       xdescent, ydescent, width, height, fontsize,
                       trans):

        # enlarge the image by these margins
        sx, sy = self.image_stretch 

        # create a bounding box to house the image
        bb = Bbox.from_bounds(xdescent - sx,
                              ydescent - sy,
                              width + sx,
                              height + sy)

        tbb = TransformedBbox(bb, trans)
        image = BboxImage(tbb)
        image.set_data(self.image_data)

        self.update_prop(image, orig_handle, legend)

        return [image]

    def set_image(self, image_path, image_stretch=(0, 0)):
        if not os.path.exists(image_path):
            # sample = ("grace_hopper.png", asfileobj=False)
            self.image_data = img.imread("home.png")
        else:
            self.image_data = img.imread("home.png")

        self.image_stretch = image_stretch

# random data
x = np.random.randn(100)
y = np.random.randn(100)
y2 = np.random.randn(100)

# plot two series of scatter data
s = plt.scatter(x, y, c='b')
s2 = plt.scatter(x, y2, c='r')

# setup the handler instance for the scattered data
custom_handler = ImageHandler()
custom_handler.set_image("home.png",
                         image_stretch=(0, 20)) # this is for grace hopper

# add the legend for the scattered data, mapping the
# scattered points to the custom handler
plt.legend([s, s2],
           ['Scatters 1', 'Scatters 2'],
           handler_map={s: custom_handler, s2: custom_handler},
           labelspacing=2,
           frameon=False)
```

```python
import matplotlib.pyplot as plt
from matplotlib.transforms import TransformedBbox
from matplotlib.image import BboxImage, Bbox
from matplotlib.legend_handler import HandlerBase
# from matplotlib._png import read_png
import pandas as pd
import os

class ImageHandler(HandlerBase):
    def create_artists(self, legend, orig_handle,
                       xdescent, ydescent, width, height, fontsize,
                       trans):

        # enlarge the image by these margins
        sx, sy = self.image_stretch 

        # create a bounding box to house the image
        bb = Bbox.from_bounds(xdescent - sx,
                              ydescent - sy,
                              width + sx,
                              height + sy)

        tbb = TransformedBbox(bb, trans)
        image = BboxImage(tbb)
        image.set_data(self.image_data)

        self.update_prop(image, orig_handle, legend)

        return [image]

    def set_image(self, image_path, image_stretch=(0, 0)):
        if not os.path.exists(image_path):
            # sample = ("grace_hopper.png", asfileobj=False)
            self.image_data = img.get_sample_data("grace_hopper.png")
        else:
            self.image_data = img.imread("home.png")

        self.image_stretch = image_stretch


fig = plt.figure(figsize=(10, 6), facecolor='white')
ax = fig.add_subplot(1, 1, 1)
# Les deux lignes ci-dessus permettent plus d'options que le simple
# fig, ax = plt.subplots()

work_df = sub_epi_df
deads = plt.bar(work_df.event, work_df.deaths)

arr_img= img.imread("home.png")

units_per_icon = 5000000
taille = 1

# imagebox.image.axes = ax
for x,y in zip(work_df.event, work_df.deaths):
    print("x, y", x, y )
    nb_of_icons = int(round(y / units_per_icon))
    print("nb_incones", nb_of_icons)
    for i in range(nb_of_icons):

        imagebox = OffsetImage(arr_img, zoom=0.7* math.sqrt(taille))
        ab = AnnotationBbox(imagebox, (x, int(y/ nb_of_icons) * i + (y/nb_of_icons/2)),
                            xybox=(0,  0),
                            xycoords='data',  # on s'exprime en coordonnées de data
                            boxcoords="offset points",  # le décalage est en points
                            frameon=False,  # on élimine le cadre autour du dessin.
                            pad=1,
                        # arrowprops=dict(
                        #     arrowstyle="->",
                        #     connectionstyle="angle,angleA=0,angleB=90,rad=3")
                            )
        ax.add_artist(ab)



# setup the handler instance for the scattered data
custom_handler = ImageHandler()
custom_handler.set_image("home.png",
                         image_stretch=(0, 20)) # this is for grace hopper

# add the legend for the scattered data, mapping the
# scattered points to the custom handler
plt.legend([s, s2],
           ['Scatters 1', 'Scatters 2'],
           handler_map={s: custom_handler, s2: custom_handler},
           labelspacing=2,
           frameon=False)
plt.legend([deads],
           ["1 M Morts"], 
           frameon = False, 
           handler_map={deads:custom_handler} )


plt.show()
```
